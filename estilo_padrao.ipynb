```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set a base style
sns.set_style("whitegrid")  # Options: darkgrid, whitegrid, dark, white, ticks

# Customize the theme
sns.set_context("talk", font_scale=1)  # Set font size for labels
sns.set_palette("pastel")  # Set a color palette

# Example plot
sns.lineplot(x=[1, 2, 3], y=[4, 5, 6])
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Seaborn Custom Styled Plot')
plt.show()
```
